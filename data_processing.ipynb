```python
import numpy as np

import matplotlib.pyplot as plt

from lambeq import NumpyModel, QuantumTrainer, BinaryCrossEntropyLoss, SPSAOptimizer
```

```python
def gen_plot(n_layers,N,ansatz):
    logdir='2systematic_runs\Epochs_2000--A_0.05--N_{}--S_1--L_{}--Ansatz_{}'.format(N,n_layers,ansatz)
    seeds=[0, 10, 50, 77, 100, 111, 150, 169, 200, 234, 250, 300, 350, 400, 450]
    models=[]
    trainers=[]
    for i in seeds:
        model=NumpyModel.from_checkpoint(logdir+'\Seed_{}\model.lt'.format(i))
        models.append(model)
        trainers.append(QuantumTrainer(
            model,
            loss_function=BinaryCrossEntropyLoss(),
            epochs=2000,
            optimizer=SPSAOptimizer,
            optim_hyperparams={'a': 0.01, 'c': 0.06, 'A':0.01*2000},
            evaluate_functions={'acc': lambda y_hat, y: np.sum(np.round(y_hat) == y) / len(y) / 2 },
            from_checkpoint=True,
            log_dir=logdir+'\Seed_{}'.format(i)
        )
        )
        train_costs=[]
        train_eval=[]
        val_costs=[]
        val_eval_results=[]

        for trainer in trainers:
            train_costs.append(trainer.train_epoch_costs)
            train_eval.append(trainer.train_eval_results['acc'])
            val_costs.append(trainer.val_costs)
            val_eval_results.append(trainer.val_eval_results['acc'])

    train_costs=np.array(train_costs)
    train_eval=np.array(train_eval)
    val_costs=np.array(val_costs)
    val_eval_results=np.array(val_eval_results)

    train_costs_mean=np.mean(train_costs,axis=0)
    train_eval_mean=np.mean(train_eval,axis=0)
    val_costs_mean=np.mean(val_costs,axis=0)
    val_eval_results_mean=np.mean(val_eval_results,axis=0)

    return train_costs_mean, train_eval_mean, val_costs_mean, val_eval_results_mean
    
```

```python
def single_model_plot(n_layers,N,ansatz):
    logdir='2systematic_runs\Epochs_2000--A_0.05--N_{}--S_1--L_{}--Ansatz_{}'.format(N,n_layers,ansatz)
    seeds=[0, 10, 50, 77, 100, 111, 150, 169, 200, 234, 250, 300, 350, 400, 450]
    trainers=[]
    for i in seeds:
        model=NumpyModel.from_checkpoint(logdir+'\Seed_{}\model.lt'.format(i))
        trainers.append(QuantumTrainer(
            model,
            loss_function=BinaryCrossEntropyLoss(),
            epochs=2000,
            optimizer=SPSAOptimizer,
            optim_hyperparams={'a': 0.01, 'c': 0.06, 'A':0.01*2000},
            evaluate_functions={'acc': lambda y_hat, y: np.sum(np.round(y_hat) == y) / len(y) / 2 },
            from_checkpoint=True,
            log_dir=logdir+'\Seed_{}'.format(i)
        )
        )
        train_costs=[]
        train_eval=[]
        val_costs=[]
        val_eval_results=[]

        for trainer in trainers:
            train_costs.append(trainer.train_epoch_costs)
            train_eval.append(trainer.train_eval_results['acc'])
            val_costs.append(trainer.val_costs)
            val_eval_results.append(trainer.val_eval_results['acc'])

    train_costs=np.array(train_costs)
    train_eval=np.array(train_eval)
    val_costs=np.array(val_costs)
    val_eval_results=np.array(val_eval_results)

    train_costs_mean=np.mean(train_costs,axis=0)
    train_eval_mean=np.mean(train_eval,axis=0)
    val_costs_mean=np.mean(val_costs,axis=0)
    val_eval_results_mean=np.mean(val_eval_results,axis=0)

    fig, ((ax_tl, ax_tr), (ax_bl, ax_br)) = plt.subplots(2, 2, sharex=True, sharey='row', figsize=(10, 6))
    ax_tl.set_title('Training set')
    ax_tr.set_title('Development set')
    ax_bl.set_xlabel('Iterations')
    ax_br.set_xlabel('Iterations')
    ax_bl.set_ylabel('Accuracy')
    ax_tl.set_ylabel('Loss')


    colours = iter(plt.rcParams['axes.prop_cycle'].by_key()['color'])
    range_ = np.arange(1, 2000 + 1)
    ax_tl.plot(range_, train_costs_mean, color='k')#next(colours))
    for trainer in trainers:
        ax_tl.plot(range_,trainer.train_epoch_costs,color='r',alpha=0.1)

    ax_bl.plot(range_, train_eval_mean, color='k')#next(colours))
    for trainer in trainers:
        ax_bl.plot(range_,trainer.train_eval_results['acc'],color='r',alpha=0.1)

    ax_tr.plot(range_, val_costs_mean, color='k')#next(colours))
    for trainer in trainers:
        ax_tr.plot(range_,trainer.val_costs,color='r',alpha=0.1)


    ax_br.plot(range_, val_eval_results_mean, color='k')#next(colours))
    for trainer in trainers:
        ax_br.plot(range_,trainer.val_eval_results['acc'],color='r',alpha=0.1)


    plt.suptitle('N: {}, S:1, L:{}, Ansatz:{}'.format(N,n_layers,ansatz))
    plt.savefig('figures\single_model\{}\Epochs_2000--A_0.05--N_{}--S_1--L_{}.png'.format(ansatz,N,n_layers))

    
```

```python
import re
from statistics import fmean
from collections import defaultdict
```

```python
def get_values(N,L,seed_no,Base):
    if Base:
        file_path=f'runs\Baseline\Epochs_2000--A_0.05--N_{N}--S_1--L_{L}--Ansatz_Sim15Ansatz\Seed_{seed_no}\info_file.txt'
    else:
        file_path=f'runs\Proper\Epochs_2000--A_0.05--N_{N}--S_1--L_{L}--Ansatz_FslBase\Seed_{seed_no}\info_file.txt'
    
    # Dictionary to store the extracted values
    results = {}

    # Open and read the file
    with open(file_path, 'r') as file:
        content = file.read()

        # Define the patterns to search for each required value
        patterns = {
            'Test accuracy': r'Test accuracy:\s*([0-9.]+)',
            'Test accuracy best model': r'Test accuracy best model:\s*([0-9.]+)',
            'OOV test accuracy': r'OOV test accuracy:\s*([0-9.]+)',
            'OOV test accuracy best model': r'OOV test accuracy best model:\s*([0-9.]+)',
            'Redundant test accuracy': r'Redundant test accuarcy:\s*([0-9.]+)',
            'Redundant test accuracy best model': r'Redundant test accuracy best model:\s*([0-9.]+)'
        }

        # Extract the values using regular expressions
        for key, pattern in patterns.items():
            match = re.search(pattern, content)
            if match:
                results[key] = float(match.group(1))

    
    return results
```

```python
def get_average(N,L,seeds,Base):

    all_results=defaultdict(list)
    for seed in seeds:
        try:
            result_dict=get_values(N=N,L=L,seed_no=seed,Base=Base)
            for key, value in result_dict.items():
                all_results[key].append(value)
        except Exception as e:
            print(e)
    averages={}
    for key, value in all_results.items():
        averages[key]=fmean(value)

    maximum=max(all_results['OOV test accuracy'])
    return averages, maximum
```

```python
import numpy as np
preq_embeddings={}
with open("resources/embeddings/common_crawl/glove.42B.300d.txt", 'r', encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "double")
        preq_embeddings[word] = vector
```

```python

def read_data(filename):
    labels, sentences = [], []
    with open(filename) as f:
        for line in f:
            t = int(line[0])
            labels.append([t, 1-t])
            sentences.append(line[1:].strip())
    return labels, sentences
```

```python
def get_unique_words(sentences):
    unique_words = set()
    
    for sentence in sentences:
        words = sentence.split()
        for word in words:
            # Remove punctuation and convert to lower case
            clean_word = ''.join(char for char in word if char.isalnum()).lower()
            if clean_word:
                unique_words.add(clean_word)
    
    return unique_words
```

```python
train_labels, train_data = read_data('resources/dataset/new_mc_train_data.txt')
dev_labels, dev_data = read_data('resources/dataset/new_mc_dev_data.txt')
test_labels, test_data = read_data('resources/dataset/new_mc_test_data_seen.txt')
OOV_test_labels, OOV_test_data = read_data('resources/dataset/new_mc_test_data_OOV.txt')
redundant_test_labels, redundant_test_data = read_data('resources/dataset/new_mc_test_data_redundancy.txt')
```

```python
all_sentences=train_data+dev_data+test_data+OOV_test_data+redundant_test_data
unique_words=get_unique_words(all_sentences)
unique_words.remove('1')
unique_words=list(unique_words)
```
